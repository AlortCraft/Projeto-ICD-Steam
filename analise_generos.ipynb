```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
df_games_info = pd.read_csv("data/all_games_info.csv")
df_games_info
```

```python
generos = df_games_info.genres
all_generos = []
for ges in generos:
    ges = ges.split(", ")
    for ge in ges:
        all_generos.append(ge)

all_generos
```

```python
plt.hist(all_generos, bins = 40)
plt.title("Numero de jogos com cada genero")
plt.xlabel("Generos")
plt.ylabel("Numero de jogos")
plt.xticks(rotation=-80)
plt.show()
```

```python
dfs_played_years = [
    pd.read_csv("data/most_played_2019.csv"),
    pd.read_csv("data/most_played_2020.csv"),
    pd.read_csv("data/most_played_2021.csv"),
    pd.read_csv("data/most_played_2022.csv"),
    pd.read_csv("data/most_played_2023.csv"),
    pd.read_csv("data/most_played_2024.csv")
    ]


df_all = pd.concat(dfs_played_years).groupby("id").mean()
df_all = pd.merge(df_games_info, df_all, on="id")
df_all

df_review = pd.read_csv("data/all_games_review.csv")
df_all = pd.merge(df_all, df_review, on="id")
df_all
```

```python
df_played_2019 = pd.read_csv("data/most_played_2019.csv")
df_review = pd.read_csv("data/all_games_review.csv")

df_2019_merged = pd.merge(df_games_info, df_played_2019, on="id")
df_2019_merged = pd.merge(df_2019_merged, df_review, on="id")
df_2019_merged["genres"] = df_2019_merged["genres"]

df_2019_merged["gpu_vram"].replace("Unknown", np.nan)
df_2019_merged["gpu_vram"] = df_2019_merged["gpu_vram"][df_2019_merged["gpu_vram"].notna()]

df_2019_merged
```

```python
all_generos_set = set(all_generos)

media_generos = {}
media_vram = {}
media_preco = {}
media_score = {}

std_generos = {}
for gen in all_generos_set:
    filtro = df_all["genres"].str.contains(gen)
    
    df_all_medias = df_all.loc[filtro, "average"]
    df_all_gpus = df_all.loc[filtro, "gpu_vram"]
    df_all_precos = df_all.loc[filtro, "price_USD"]
    df_all_score = df_all.loc[filtro, "score"]
    
    media_generos[gen] = df_all_medias.mean()
    media_vram[gen] = df_all_gpus.mean()
    media_preco[gen] = df_all_precos.mean()
    media_score[gen] = df_all_score.mean()
    
    std_generos[gen] = df_all_medias.std()
    
    
```

```python
plt.bar(media_generos.keys(), media_generos.values())
plt.title("Generos X media de jogadores")
plt.xlabel("Generos")
plt.ylabel("Media de jogadores simultaneamente")
plt.xticks(rotation=-80)
plt.margins(x=.08)
plt.show()
```

```python
plt.bar(std_generos.keys(), std_generos.values())
plt.title("Generos X Desvio padrao entre media de jogadores de cada ano")
plt.xlabel("Generos")
plt.ylabel("Desvio padrao da media de jogadores simultaneamente")
plt.xticks(rotation=-80)
plt.margins(x=.08)
plt.show()
```

```python
cont_generos = Counter(all_generos)

media_por_jogos = {}
for gen, jogadores in media_generos.items():
    media = jogadores / cont_generos[gen]
    media_por_jogos[gen] = media
    
plt.bar(media_por_jogos.keys(), media_por_jogos.values())
plt.title("Media de Jogadores por jogo de cada genero")
plt.xlabel("Generos")
plt.ylabel("Media de jogadores por jogo")
plt.xticks(rotation=-80)
plt.show()
```

```python


data = list(media_generos.values())

# Remove any NaN values
clean_data = [value for value in data if not np.isnan(value)]

plt.boxplot(clean_data)
plt.show()
```
